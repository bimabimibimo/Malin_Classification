```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('fifa20.csv')
df.head()
```

```python
data = pd.DataFrame()
```

```python
data['height_cm']=df['height_cm']
data['weight_kg']=df['weight_kg']
data['body_type']=df['body_type']
```

```python
print(data.body_type.unique())
print(data.)
```

```python
data = data[data.body_type != 'Messi']
data = data[data.body_type != 'C. Ronaldo']
data = data[data.body_type != 'Neymar']
data = data[data.body_type != 'PLAYER_BODY_TYPE_25']
data = data[data.body_type != 'Courtois']
data = data[data.body_type != 'Shaqiri']
data = data[data.body_type != 'Akinfenwa']
data.shape
```

```python
X = data.iloc[:, :-1]
Y = data.iloc[:, -1]
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=50)
```

```python
from sklearn.svm import SVC
classifier = SVC( kernel = 'rbf', C=200, gamma = 200)
classifier.fit(X_train, y_train)
```

```python
predict = classifier.predict(X_test)
```

```python
predict
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix
coma = confusion_matrix(y_test, predict)
```

```python
coma.shape
```

```python
from sklearn.metrics import precision_recall_fscore_support
precision_recall_fscore_support(y_test,predict, average='weighted')
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, predict)
```
